```python
x= np.linspace(0,5,50)
x1 = np.linspace(0,10,50)
x2 = np.linspace(0,15,50)
x3 = np.linspace(0,20,50)
x4 = np.linspace(0,25,50)
```

```python
y = np.sin(x)
y1 = np.sin(x1)
y2 = np.cos(x2)
y3 = np.sin(x3)
y4 = np.cos(x4)
```

```python
fig , ax = plt.subplots()

ax.plot(x1,y1,label='Sin(x)',color = 'b',linestyle='--',linewidth=2)
ax.plot(x2,y2,label = 'Cos(x)',color='r',linestyle='-',linewidth=2)

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

ax.set_xlim([0,10])
ax.set_ylim([-1.2,1.2])
#custom line style rather than ['-','--','-.',':']
ax.grid(True,linestyle=(0,(10,2,10,2)),alpha=0.5)

ax.text(5,1,'Maximum',ha='center',va='center',color = 'green',fontsize=12)

ax.annotate('Inflection Point',xy=(np.pi/2,0),xycoords='data',xytext=(3,0.8),arrowprops=dict(facecolor='black',arrowstyle='->'),fontsize=10)

ax.axvline(x=np.pi/2,color='grey',linestyle='-.',linewidth=1)
ax.axhline(y=0,color='grey',linestyle='-.',linewidth=1)

ax.set_xticks([0,np.pi/2,np.pi,3*np.pi/2,4*np.pi/2,5*np.pi/2])
ax.set_xticklabels(['0','$\pi/2$','$\pi$','$3\pi/2$','$2\pi$','$5\pi/2$'])

plt.savefig('plot.png',dpi=1000)
plt.legend()

plt.show
```

```python
#qualitative color scale used for nominal categorail varable

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import random

fig,ax = plt.subplots()

x = [np.linspace(0,np.pi,50),np.linspace(0,2*np.pi,50)]
y = [np.sin(x[0]),np.sin(x[1])]

cmap = cm.get_cmap('YlGnBu')
#z = np.random.randint(1,cmap.N)
z=[cmap.N/250,cmap.N/100]

for i in range(len(x)):
    color = cmap(z[i])
    ax.plot(x[i],y[i],color = color)
    

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Qualitative Color Scale')
```

### Grouped Bar Chat

```python
categories = ['A', 'B', 'C', 'D']
values1 = [10, 15, 7, 12]
values2 = [8, 12, 9, 10]

bar_wid=.3
barposition_1 = np.arange(len(categories))
barposition_2 = barposition_1+bar_wid

fig,ax = plt.subplots(1,2,figsize=(15,5))

ax[0].bar(barposition_1,values1,label='Group1',color='r',width=bar_wid)
ax[0].bar(barposition_2,values2,label='Group2',color='b',width=bar_wid)
ax[0].set_xlabel('Categories')
ax[0].set_ylabel('Values')

ax[0].set_xticks(barposition_1+bar_wid/2)
ax[0].set_xticklabels(categories)
plt.legend()


bar_positions = np.arange(len(categories))
ax[1].bar(bar_positions, values1, label='Group 1')
ax[1].bar(bar_positions, values2, bottom=values1, label='Group 2')
ax[1].set_xlabel('Categories')
ax[1].set_ylabel('Values')
ax[1].set_xticks(bar_positions)
ax[1].set_xticklabels(categories)
```

```python
import plotly.express as px
import numpy as np

# Generate some example data
data = np.random.randn(100)

# Calculate the empirical cumulative distribution function (ECDF)
sorted_data = np.sort(data)
ecdf = np.arange(1, len(sorted_data) + 1) / len(sorted_data)

# Create an EDCF plot using Plotly Express
fig = px.line(x=sorted_data, y=ecdf, labels={'x': 'Data', 'y': 'ECDF'})

# Set plot layout
fig.update_layout(
    title='Empirical Distribution Cumulative Function (EDCF) Plot',
    xaxis_title='Data',
    yaxis_title='ECDF'
)

# Show the plot
fig.show()
```

## QQ plot

```python
from scipy.stats import stats


import scipy.stats as stats

fig, ax = plt.subplots()
stats.probplot(data, dist='norm', plot=ax)

```

```python
import plotly.express as px
import numpy as np
import statsmodels.api as sm
#api generates theoretical quantiles and sample quantiles for QQ plot

data = np.random.randn(100)


theoretical_quantiles = sm.ProbPlot(data).theoretical_quantiles


sample_quantiles = sm.ProbPlot(data).sample_quantiles


fig = px.scatter(x=theoretical_quantiles, y=sample_quantiles, labels={'x': 'Theoretical Quantiles', 'y': 'Sample Quantiles'})

line = go.Scatter(x=[theoretical_quantiles.min(), theoretical_quantiles.max()],
                  y=[theoretical_quantiles.min(), theoretical_quantiles.max()],
                  mode='lines',
                  line=dict(color='red'),
                  showlegend=False)
fig.add_trace(line)

fig.update_layout(
    title='Quantile-Quantile (Q-Q) Plot',
    xaxis_title='Theoretical Quantiles',
    yaxis_title='Sample Quantiles',
    title_x=0.5,
    title_y=1
)

fig.show()
```
