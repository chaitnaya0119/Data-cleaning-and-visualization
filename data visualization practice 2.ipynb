## scatter plot with regression line
* It shows the regression line with individual points


```python
x = np.arange(0,10,0.1)
y= 2*x + np.random.randn(len(x))
fig,ax = plt.subplots()
ax.scatter(x,y)
m,c= np.polyfit(x,y,1)
ax.plot(x,m*x+c,color='red')
```

## Moving Average Plot

```python
import numpy as np
import matplotlib.pyplot as plt
time = np.arange(0,10,0.1)
data = np.sin(time)+np.random.randn(len(time))*0.2
window = 5
moving_avg = np.convolve(data,np.ones(window)/window,mode='valid')
time_ma = time[window-1:]
fig,ax = plt.subplots()
ax.plot(time,data,label='Original')
ax.plot(time_ma,moving_avg,label='Moving Average')
ax.set_xlabel('Time')
ax.set_ylabel('Value')
ax.set_title('Trend with Moving Average')
ax.legend()
```

```python
time = np.arange(0,10,0.1)
data1 = np.sin(time)
data2 = np.cos(time)
fig,ax = plt.subplots()
ax.fill_between(time,data1,alpha=0.5,label='Component 1')
ax.fill_between(time,data2,alpha=0.5,label='Component 2')
plt.legend()
```

```python
import plotly.graph_objs as go


trace = go.Scatter(
    x=[1, 2, 3, 4, 5],
    y=[10, 20, 30, 40, 50],
    mode='markers+lines',
    marker=dict(
        color='blue'
    ),
    error_y=dict(
        type='data',
        array=[1, 2, 3, 4, 5],
        visible=True
    ),
    error_x=dict(
        type='data',
        array=[0.5, 1, 1.5, 2, 2.5],
        visible=True
    )
)


fig = go.Figure(data=[trace])


fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
fig ,ax = plt.subplots()
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = np.random.randint(20, 200, size=50)
ax.scatter(x, y, c=colors, s=sizes, alpha=0.8, cmap='viridis')
```

```python
import plotly.graph_objs as go

# Create a scatter plot with custom sizes and colors
trace = go.Scatter(
    x=[1, 2, 3, 4, 5],
    y=[10, 20, 30, 40, 50],
    mode='markers',
    marker=dict(
        size=[10, 20, 30, 40, 50],
        color=['red', 'green', 'blue', 'orange', 'purple'],
        line=dict(
            width=2,
            color='black'
        )
    )
)

# Create a figure object
fig = go.Figure(data=[trace])

# Show the figure
fig.show()
```
