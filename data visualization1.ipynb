## Visualizing Many Distributions at Once

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python

data1 = np.random.normal(0, 1, 1000)
data2 = np.random.normal(2, 0.5, 1000)
data3 = np.random.normal(-1, 2, 1000)

fig,ax = plt.subplots()
ax.hist(data1,label='Distributiion 1',alpha=0.5,bins=30)
ax.hist(data2,label='Distributiion 2',alpha=0.5,bins=30)
ax.hist(data3,label='Distributiion 3',alpha=0.5,bins=30)

plt.legend()
```

```python
import plotly.express as px
import numpy as np

data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 1
data3 = np.random.randn(1000) - 1


fig = px.histogram(data1, title='Distribution 1', opacity=1, nbins=30)


fig.add_histogram(x=data2, opacity=0.2)


fig.add_histogram(x=data3, opacity=0.3)

fig.update_traces(name='Distribution 2', selector=dict(name='<extra>'))
fig.update_traces(name='Distribution 3', selector=dict(name='<extra>'))

fig.update_layout(
    title='Multiple Histograms in the Same Plot',
    xaxis_title='Value',
    yaxis_title='Frequency'
)

# Show the plot
fig.show()
```

```python

data1 = np.random.normal(0, 1, 1000)
data2 = np.random.normal(2, 0.5, 1000)
data3 = np.random.normal(-1, 2, 1000)
data=[data1,data2,data3]
plt.violinplot(data,showmedians=True)

plt.xticks([1,2,3],['Distribution 1', 'Distribution 2', 'Distribution3'])
ax.set_ylabel('Values')
ax.set_title('Violin Plots')
```

```python

import plotly.graph_objects as go



fig = go.Figure()

fig.add_trace(go.Violin(y=data1, box_visible=True, line_color='blue', name='data1'))
fig.add_trace(go.Violin(y=data2, box_visible=True, line_color='green', name='data2'))
fig.add_trace(go.Violin(y=data3, box_visible=True, line_color='red', name='data3'))


fig.update_layout(
    xaxis = dict(
        tickmode = 'array',
        tickvals = [0, 1, 2],
        ticktext = ['Data 1', 'Data 2', 'Data 3']
    )
)

fig.show()




```
